<h1>Drop the not useful column for dataset</h1>

```python
df.drop(['Not_Useful_Column'], axis= 'columns', inplace=True)
```

```python
df
```

<h1>Checking for duplicates and drop them if exist</h1>

```python
df[df.duplicated]
```

<h4> after checking for duplicates we can see that record number 20 is duplicated hence we remove it </h4>

```python
df.drop_duplicates(inplace = True)
```

```python
df
```

<h1>Removing special characters</h1>

```python
df['Last_Name'] = df['Last_Name'].str.replace(r'[!@#$%^&*./_-]', '', regex=True)
df
```

<h1>Cleaning phone number</h1>

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-z A-Z 0-9]','', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6]+'-'+ x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(nan--|Na--)','',regex=True)
df
```

<h1>Getting rid of N/a</h1>

```python
df.replace('N/a','',inplace=True)
df
```

```python
df.fillna('', inplace=True)
df
```

<h1>Reformatting the last two columns</h1>

```python
df['Paying Customer']= df['Paying Customer'].str.replace(r'No|N' ,'No', regex=True).replace(r'Yes|Y', 'Yes',regex=True)
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace(r'No|N' ,'No', regex=True).replace(r'Yes|Y', 'Yes',regex=True)
df
```

<h1>Replacing the empty records</h1>

<h3>Now we can deal with updated data as we wish. We can either extract the empty records and let customers fill them, or we can fill the last two records 'Paying Cutomer" and "Do_Not_Contact" using mode method </h3>

<h3>We will fill the last two columns and extract customers who has their phone number columns missing to come and fill it up </h3>

```python
df['Paying Customer'].fillna(df['Paying Customer'].mode()[0])
```

<h3>Sine we have converted all NaN value to string then we can't use the mode method so we have to fill them again with NaN</h3>

```python
df['Paying Customer']= df['Paying Customer'].replace('', pd.NA)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', pd.NA)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].fillna(df['Paying Customer'].mode()[0])
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna(df['Do_Not_Contact'].mode()[0])
df
```

<h1>Set a new table for empty records to be filled</h1>

```python
df_empty = pd.DataFrame(df)
df_empty.to_excel('uncleaned_data.xlsx')
df_empty
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x, inplace=True)
for y in df_empty.index:
    if df_empty.loc[y,'Phone_Number'] != '':
        df_empty.drop(y, inplace=True)
```
